```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([text])
```

```python
vocabulary=tokenizer.word_index
```

```python
vocabulary
```

```python
vocabulary_size=len(vocabulary)+1
```

```python
vocabulary_size
```

```python
sequences=[]
for sentence in text.split("\n"):
    tokenized_sen=tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1,len(tokenized_sen)):
        n_gram_sequence=tokenized_sen[:i+1]
        sequences.append(n_gram_sequence)
```

```python
sequences
```

```python
max_seq_len=max([len(seq) for seq in sequences])
```

```python
max_seq_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
paded_sequences=pad_sequences(sequences,maxlen=max_seq_len,padding='pre')
```

```python
paded_sequences
```

```python
X=paded_sequences[:,:-1]
y=paded_sequences[:,-1]
```

```python
X
```

```python
y
```

```python
len(y)
```

```python
y.shape
```

```python
#Convert y to catergorical y (One Hot Encoding)
from tensorflow.keras.utils import to_categorical   
y=to_categorical(y,num_classes=vocabulary_size)
```

```python
y.shape
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(input_dim=vocabulary_size,output_dim=200,input_shape=(max_seq_len-1,)))
model.add(LSTM(150))
model.add(Dense(vocabulary_size,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
input="kya"

tokenized_input=tokenizer.texts_to_sequences([input])[0]
print(tokenized_input)
paded_tokenized_input=pad_sequences([tokenized_input],maxlen=max_seq_len-1,padding='pre')
print(paded_tokenized_input)
val=np.argmax(model.predict(paded_tokenized_input))
print(val)

for word, index in tokenizer.word_index.items():
    if index==val:
        print("Next word is:",word)
```
